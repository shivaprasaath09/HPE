```
import pandas as pd
#feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```
train_df = pd.read_csv("/dbfs/FileStore/shared_uploads/shiva.prasaath@bajajfinserv.in/Phising_Training_Dataset-1.csv")
test_df =  pd.read_csv("/dbfs/FileStore/shared_uploads/shiva.prasaath@bajajfinserv.in/Phising_Testing_Dataset.csv")
```

```
train_df
```

```
train_df = train_df.replace(-1,3)
test_df = test_df.replace(-1,3)
```

```
train_output = train_df["Result"]
train_df = train_df.drop(["Result"],axis = 1)
```

```
#Chi square test for feature selection 

chi2_features = SelectKBest(chi2, k = 8)
X_kbest_features = chi2_features.fit_transform(train_df.values, train_output)
```

```
names = train_df.columns.values[chi2_features.get_support()]
train_df = train_df[names]
```

```
test_df = test_df[names]
```

```
#Random forest classifier 
X, y = make_classification(n_samples=1000, n_features=4,n_informative=2, n_redundant=0,random_state=0, shuffle=False)
```

```
clf = RandomForestClassifier(max_depth=8 , random_state=0)
```

```
clf.fit(train_df, train_output)
```

```
y_pred = clf.predict(test_df)
```
